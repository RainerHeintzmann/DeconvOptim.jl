```julia
using Pkg
Pkg.activate(".")
```

```julia
using Revise
using DeconvOptim, TestImages, ImageShow, Plots, LinearAlgebra, Noise
using FourierTools, FFTW, Images, Tullio
using Statistics, StatsBase
```

```julia
img = Float32.(testimage("resolution_test_512"));
radius = 51.2
f_max = 51.2 / size(img, 1) * 2
psf = Float32.(DeconvOptim.generate_psf(size(img), radius));
img_n = Float32.(poisson(DeconvOptim.conv(img, psf), 100));
```

```julia
iter = 15
```

```julia
@time img_gauss, r = deconvolution(img_n, psf, loss=Gauss(),regularizer=nothing, iterations=iter)
```

```julia
@time img_pois, r = deconvolution(img_n, psf, loss=Poisson(), regularizer=nothing, iterations=iter)
```

```julia
@time img_ansc, r = deconvolution(img_n, psf, loss=Anscombe(3//8),regularizer=nothing, iterations=iter)
```

```julia
@time img_scagauss, r = deconvolution(img_n, psf, loss=ScaledGauss(), regularizer=nothing, iterations=iter)
```

```julia
#[Gray.(img_gauss) Gray.(img_pois) Gray.(img_scagauss) Gray.(img_ansc)]
```

```julia
g(x) = begin
    a, b = DeconvOptim.relative_energy_regain(img, x)
    a ./= f_max
    return (a,b)
end

plot(g(img_gauss), label="gauss", xlim=(0, 3), ylim=(-1, 1), 
     ylabel="relative energy regain", xlabel="\$ f / f_\\textrm{max}\$")
plot!(g(img_pois), label="Poisson")
plot!(g(img_scagauss), label="Scaled Gauss")
plot!(g(img_ansc), label="Anscombe")
```
